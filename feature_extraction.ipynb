```python
import torch
from src.models.image_encoder import ImageEncoder

device = "cuda" if torch.cuda.is_available() else "cpu"

img_path = "data/raw/posters/1.jpg"

image_encoder = ImageEncoder(device=device)

features = image_encoder([img_path])

print("Output shape:", features.shape)
print("Sample values:", features[0][:10])
print("Contains NaN:", torch.isnan(features).any().item())
```

```python
from src.models.text_encoder import TextEncoder
text_encoder = TextEncoder(device=device)

# sample
texts = [
    "test text to see",
    "test text two"
]

embeddings = text_encoder(texts)
print("Shape:", embeddings.shape)
print("First vector sample:", embeddings[0][:10])
print("NaN check:", torch.isnan(embeddings).any().item())
```

```python
from src.models.early_fusion_feature_extraction import EarlyFusionFeatureExtractor

combined_feature_extractor = EarlyFusionFeatureExtractor(
    image_encoder=image_encoder,
    text_encoder=text_encoder
)
img_path = ["data/raw/posters/1.jpg"]

text = ['Example text about Toy Story']

combined_features = combined_feature_extractor(image_paths=img_path, texts=text)

print("Fused feature shape:", combined_features.shape)
```
