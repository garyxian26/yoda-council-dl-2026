```python
train_ds = MovieLensMultiModal(
    root=DATA_ROOT,
    split="train",
    split_method="temporal_ratio",
    require_poster=True,
    min_user_interactions=5,
    min_item_interactions=5,
)

val_ds = MovieLensMultiModal(
    root=DATA_ROOT,
    split="val",
    split_method="temporal_ratio",
    require_poster=True,
    min_user_interactions=5,
    min_item_interactions=5,
)

test_ds = MovieLensMultiModal(
    root=DATA_ROOT,
    split="test",
    split_method="temporal_ratio",
    require_poster=True,
    min_user_interactions=5,
    min_item_interactions=5,
)

print(len(train_ds), len(val_ds), len(test_ds))
print(train_ds[0].keys())
```
